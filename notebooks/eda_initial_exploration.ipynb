```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load raw data

df = pd.read_csv("../data/raw/eeg_eye_state.csv")
print("dataset loaded successfully. shape:",df.shape)
```

```python
# Preview Data
print(df.head)
```

```python
# Check missing values

print("Missing value:")
print(df.isnull().sum())
```

```python
# Summary
print(df.describe())
```

```python
# Distribution of target

print("Target Distribution:")
print(df['eyeDetection'].value_counts())
```

```python
# Plot Distribution

sns.countplot(data=df , x = 'eyeDetection')
plt.title('Distribution of Eye State')
plt.show()
```

```python
# Histograms

df.iloc[:,:-1].hist(figsize=(15,10))
plt.suptitle("EEG Features Distribution")
plt.show()
```

```python
# Save Cleaned Data
df.to_csv("../data/processed/eeg_clean.csv",index=False)
print("Cleaned data Saved to /data/processed/eeg_clean.csv ")
```
